## ***Import Important libraries*** 

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

### ***Load Boston House Pricing Dataset***

```python
from sklearn.datasets import load_boston
boston_df = load_boston()
```

### ***Display data***

```python
boston_df
```

```python
# Independent feature
X = pd.DataFrame(boston_df.data, columns=boston_df.feature_names)
X
```

```python
# Dependent feature
y = boston_df.target
y
```

```python
# train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state= 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_pred, y_test)
score
```

```python
## Hyperparameter Tunning
parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
  'max_features':['auto', 'sqrt', 'log2']
    
}
regressor=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
regressorcv = GridSearchCV(regressor,param_grid=parameter, cv=5, scoring='neg_mean_squared_error')
```

```python
regressorcv
```

```python
regressorcv.fit(X_train,y_train)
```

```python
regressorcv.best_estimator_
```

```python
regressorcv.best_params_
```

```python
regressorcv.best_score_
```

```python
y_pred = regressorcv.predict(X_test)
y_pred
```

```python
r2_score(y_pred,y_test)
```
