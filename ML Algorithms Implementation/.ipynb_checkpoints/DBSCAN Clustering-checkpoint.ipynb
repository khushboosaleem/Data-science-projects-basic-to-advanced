## DBSCAN Clustering

```python
# Importing DBSCAN class from sklearn.cluster for density-based clustering
from sklearn.cluster import DBSCAN

# Importing make_moons function from sklearn.datasets to generate moon-shaped data
from sklearn.datasets import make_moons

# Importing matplotlib library for data visualization
import matplotlib.pyplot as plt

# Ensuring plots are displayed inline in Jupyter Notebook
%matplotlib inline
```

```python
X,y = make_moons(n_samples=250, noise=0.05)
```

```python
X
```

```python
y
```

```python
# Creating a scatter plot of the synthetic data
# X[:, 0] represents the first feature (x-axis), and X[:, 1] represents the second feature (y-axis)
plt.scatter(X[:, 0], X[:, 1])

# Displaying the plot
plt.show()
```

```python
# Feature scaling (Standard Scaling)
# Importing the StandardScaler class from sklearn.preprocessing for standardizing features
from sklearn.preprocessing import StandardScaler

# Creating an instance of the StandardScaler class
scaler = StandardScaler()
```

```python
# Feature Scaling(Standard Scaling)
# Importing the StandardScaler class from sklearn.preprocessing for standardizing features 
from sklearn.preprocessing import StandardScaler

# Creating instance of the StandardScaler class
scaler = StandardScaler()
scaler
```

```python
# Applying standard scaling
# The fit transform method standardizes the features by remoing the mean and scaling to unit variance 

X_scaled = scaler.fit_transform(X)

# Displaying the data
X_scaled
```

## DBSCAN Algorithm


```python
# Importing the DBSCAN class from sklearn.cluster for density-based clustering
from sklearn.cluster import DBSCAN
```

```python
# Creating an instance of the DBSCAN class for density-based clustering
# Setting eps=0.5 to define the maximum distance between two samples for one to be considered as in the neighborhood of the other
dbscan = DBSCAN(eps=0.5)

# Displaying the DBSCAN instance to verify the parameters
dbscan
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
# Creating a scatter plot of the synthetic data
# X[:, 0] represents the first feature (x-axis), and X[:, 1] represents the second feature (y-axis)
# Coloring the points based on the cluster labels assigned by the DBSCAN algorithm

plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)

# Displaying the plot
plt.show()
```

```python
# Creating a scatter plot of the synthetic data
# X[:, 0] represents the first feature (x-axis), and X[:, 1] represents the second feature (y-axis)
# Coloring the points based on the original labels from the make_moons dataset

plt.scatter(X[:, 0], X[:, 1], c=y)

# Displaying the plot
plt.show()
```
