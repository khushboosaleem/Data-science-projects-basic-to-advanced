## ***Observation***

3. When Rating is 3.5 to 3.9 -----> Good
4. When Rating is 2.5 to 2.9 -----> Average
5. When Rating is 3.0 to 3.4 -----> Average
6. When Rating is 1.8 to 2.4 -----> Poor

```python
##find out which currency is used by which country?
```

```python
final_data.columns
```

```python
grouped_data = final_data.groupby(['Country','Currency']).size().reset_index(name = 'Count')
grouped_data
```

```python
len(grouped_data)
```

```python
range(len(grouped_data))
```

```python
## Crating the bar plot by using matplotlib

plt.figure(figsize = (12,6))
plt.bar(range(len(grouped_data)), grouped_data['Count'], color='skyblue')
plt.xlabel('Country - Currency')
plt.ylabel('Count')
plt.title('Count of Occurrences by Country and Currency', color = 'red')
plt.xticks(range(len(grouped_data)), [f'{country} - {currency}' for country, currency in zip(grouped_data['Country'], grouped_data['Currency'])], rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=grouped_data, x='Country', y='Count')
plt.title('Count of Occurrences by Country and Currency', color = 'red')
plt.xticks(rotation=90)
plt.xlabel('Country - Currency')
plt.ylabel('Count')
plt.show()
```

## ***Observation***

based on the data and graph indan currency has been used the most frequently.

```python
# Top 10 Cuisines
final_data['Cuisines'].value_counts()[:10]
```
